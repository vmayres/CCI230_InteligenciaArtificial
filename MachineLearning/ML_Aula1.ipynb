# Regressão Linear

```python
from sklearn import linear_model
import matplotlib.pyplot as plt

entrada = [[1.21], [1.37], [1.52], [1.58], [1.65], [1.77], [1.86],
           [1.89], [1.95], [2.10]]

peso = [42, 44, 49, 55, 59, 68, 74, 81, 85, 90]

plt.scatter(entrada, peso)
plt.xlabel("altura")
plt.ylabel("peso")
```

```python
regressao = linear_model.LinearRegression()
regressao.fit(entrada,peso)
a = regressao.coef_
b = regressao.intercept_

print("Coef Angular:", a, "Coef Linear:", b)
previsao = regressao.predict(entrada)
plt.scatter(entrada, peso, color="red")
plt.plot(entrada,previsao)
```

```python
regressao.predict([[1.85]])
```

```python
regressao.score(entrada, peso)
```

# KNN Regressão

```python
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt

# exemplos: entrada e saída
altura = [[1.75], [1.7], [1.8], [1.65], [1.83], [1.73], [1.88], [1.65],
          [1.67], [1.83]]

sapato = [41,42,43,42,43,38,44,36,38,45]
```

```python
knn = KNeighborsRegressor(n_neighbors = 5)
knn.fit(altura, sapato)
```

```python
knn.predict([[1.85]])
```

```python
knn.score(altura, sapato)
```

```python
plt.scatter(altura, sapato)
altura.sort()
previsao = knn.predict(altura)
print(previsao)
plt.plot(altura, previsao, color="red")
```

#KNN Classificação

```python
from sklearn.neighbors import KNeighborsClassifier

entrada = [[158,58], [158,59], [158, 63], [160,59], [160,60], [163,60], [163,61],
           [160,64], [163,64], [165,61], [165,62], [165,65], [168,62], [168,63],
           [168,66], [170,63], [170,64], [170, 68]]
print(len(entrada))
# saída
camiseta = ['M', 'M', 'M', 'M', 'M', 'M', 'M', 'L', 'L', 'L', 'L', 'L', 'L', 'L',
            'L', 'L', 'L', 'L']
print(len(camiseta))
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(entrada, camiseta)
```

```python
knn.predict([[163,61]])
```

```python
knn.predict_proba([[163,62]])
```

```python
print("Score:", knn.score(entrada, camiseta))
```
